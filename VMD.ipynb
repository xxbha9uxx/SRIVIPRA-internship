```python
import numpy as np
import matplotlib.pyplot as plt
from vmdpy import VMD

# Generate a sample signal
fs = 1000
T = 1
t = np.linspace(0, T, fs, endpoint=False)
s = (np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t) + 0.2 * np.sin(2 * np.pi * 20 * t))

# Parameters for VMD
alpha = 2000       # Moderate bandwidth constraint
tau = 0            # Noise-tolerance (no strict fidelity enforcement)
K = 3              # Number of modes
DC = 0             # No DC part imposed
init = 1           # Initialize omegas uniformly
tol = 1e-7         # Tolerance for convergence criterion

# Perform VMD
u, u_hat, omega = VMD(s, alpha, tau, K, DC, init, tol)

# Plot the original signal and the modes
plt.figure(figsize=(10, 8))

plt.subplot(K + 1, 1, 1)
plt.plot(t, s, 'b')
plt.title("Original Signal")

for i in range(K):
    plt.subplot(K + 1, 1, i + 2)
    plt.plot(t, u[i], 'g')
    plt.title(f"Mode {i+1}")

plt.tight_layout()
plt.show()
```
