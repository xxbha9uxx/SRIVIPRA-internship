```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA

# Generate some data
np.random.seed(0)
X = np.random.randn(100, 5)

# KPCA
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

# Plotting
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], edgecolor='k', s=50)
plt.xlabel('Kernel Principal Component 1')
plt.ylabel('Kernel Principal Component 2')
plt.title('KPCA')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(100, 2)
X[:50] = X[:50] + 5

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply KPCA with RBF kernel
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

# Plot original data
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], edgecolor='k', s=50)
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot PCA results
plt.subplot(1, 3, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], edgecolor='k', s=50)
plt.title('PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Plot KPCA results
plt.subplot(1, 3, 3)
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], edgecolor='k', s=50)
plt.title('KPCA (RBF Kernel)')
plt.xlabel('Kernel Principal Component 1')
plt.ylabel('Kernel Principal Component 2')

plt.tight_layout()
plt.show()
```
