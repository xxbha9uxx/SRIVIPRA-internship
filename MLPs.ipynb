```python
df = sns.load_dataset("Iris")
df = df.sample(frac=1)
df.head(10)
```

**Diving into Target(dependent) and Prectictor(independent) variable**

```python
# one-hot encoding
# species column is the target variable(output variable)
y = pd.get_dummies(df["species"],dtype=int)
y
```

```python
# these are the input variables (independent)
x = df.drop(["species"],axis=1)
x
```

**Splitting the dataset into training and testing dataset**

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

**Creating our Model --> MULTI-LAYER PERCEPTRON**

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python

model = Sequential()
model.add(Dense(6,activation="sigmoid"))  # Hidden layer
model.add(Dense(3,activation="softmax")) # Output layer
```

```python
model.compile(loss="categorical_crossentropy",metrics=["accuracy"])
model.fit(x_train,y_train,epochs= 25 , batch_size= 5)
```
