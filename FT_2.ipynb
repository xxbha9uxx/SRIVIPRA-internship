```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample rate and duration
sampling_rate = 1000  # Samples per second
duration = 1  # seconds
t = np.linspace(0, duration, int(sampling_rate * duration,),endpoint = False )
```

```python
# Create individual component signals
freq1 = 50  # Frequency of the first sine wave
freq2 = 120  # Frequency of the second sine wave
signal1 = np.sin(2 * np.pi * freq1 * t)
signal2 = 0.5 * np.sin(2 * np.pi * freq2 * t)
```

```python
# Combine the signals
combined_signal = signal1 + signal2
```

```python
# Compute the Fourier Transform of the combined signal
fourier_transform = np.fft.fft(combined_signal)    #output of the transform is a complex number 
#print(fourier_transform)
#fourier_transform_real = fourier_transform.real
#fourier_transform_imaginary = fourier_transform.imag
frequencies = np.fft.fftfreq(len(fourier_transform), 1/sampling_rate)
```

```python
# Plot the original signal, individual components, and Fourier Transform
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(t, signal1)
plt.title('Component Signal 1 (50 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(t, signal2)
plt.title('Component Signal 2 (120 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 3)
plt.plot(t, combined_signal)
plt.title('Combined Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 4)
plt.plot(frequencies, np.abs(fourier_transform))
plt.title('Fourier Transform of Combined Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.xlim(0, sampling_rate / 2)


plt.tight_layout()
plt.show()
```
