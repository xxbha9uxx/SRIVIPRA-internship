```python
import numpy as np
import PyWavelets as pywt
import matplotlib.pyplot as plt

# Create a sample signal
t = np.linspace(0, 1, 400, endpoint=False)
sig = np.sin(50 * 2 * np.pi * t) + np.sin(80 * 2 * np.pi * t)

# Perform the wavelet transform
coeffs = pywt.wavedec(sig, 'db1', level=6)

# Reconstruct the signal from the wavelet coefficients
reconstructed_signal = pywt.waverec(coeffs, 'db1')

# Plot the original signal and the reconstructed signal
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, sig)
plt.title('Original Signal')

plt.subplot(2, 1, 2)
plt.plot(t, reconstructed_signal[:len(t)])
plt.title('Reconstructed Signal from Wavelet Coefficients')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Create a sample signal
t = np.linspace(0, 1, 400, endpoint=False)
sig = np.sin(50 * 2 * np.pi * t) + np.sin(80 * 2 * np.pi * t)

# Perform the wavelet transform
coeffs = pywt.wavedec(sig, 'db1', level=6)

# Reconstruct the signal from the wavelet coefficients
reconstructed_signal = pywt.waverec(coeffs, 'db1')

# Plot the original signal and the reconstructed signal
plt.figure(figsize=(12, 10))

plt.subplot(3, 1, 1)
plt.plot(t, sig)
plt.title('Original Signal')

plt.subplot(3, 1, 2)
plt.plot(t, reconstructed_signal[:len(t)])
plt.title('Reconstructed Signal from Wavelet Coefficients')

# Plot individual components
plt.figure(figsize=(12, 10))
for i, coeff in enumerate(coeffs):
    plt.subplot(len(coeffs), 1, i + 1)
    plt.plot(coeff)
    plt.title(f'Component {i+1} - {"Approximation" if i == 0 else "Detail"}')

plt.tight_layout()
plt.show()
```

**CWT**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import cwt, morlet

# Create a sample signal
t = np.linspace(0, 1, 400, endpoint=False)
sig = np.sin(50 * 2 * np.pi * t) + np.sin(80 * 2 * np.pi * t)

# Define the wavelet (Morlet wavelet)
widths = np.arange(1, 128)

# Perform the Continuous Wavelet Transform (CWT)
cwt_matrix = cwt(sig, morlet, widths)

# Plot the time-frequency representation
plt.figure(figsize=(12, 6))
plt.imshow(np.abs(cwt_matrix), extent=[0, 1, 1, 128], aspect='auto', cmap='jet',
           vmax=abs(cwt_matrix).max(), vmin=0)
plt.title('Time-Frequency Representation (CWT)')
plt.ylabel('Frequency (scale)')
plt.xlabel('Time')
plt.colorbar(label='Magnitude')
plt.show()
```

```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Generate a sample signal
fs = 500  # Sampling frequency
t = np.linspace(0, 1, fs)
signal = np.sin(2 * np.pi * 7 * t) + np.sin(2 * np.pi * 13 * t)

# Add some noise
signal += 0.5 * np.random.randn(t.size)

# Perform Continuous Wavelet Transform
coefficients, frequencies = pywt.cwt(signal, scales=np.arange(1, 100), wavelet='cmor', sampling_period=1/fs)

# Plot the original signal
plt.figure(figsize=(14, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Plot the Wavelet Transform (scalogram)
plt.subplot(2, 1, 2)
plt.imshow(np.abs(coefficients), extent=[0, 1, 1, 100], cmap='jet', aspect='auto', vmax=abs(coefficients).max(), vmin=-abs(coefficients).max())
plt.title('Wavelet Transform (Scalogram)')
plt.xlabel('Time (s)')
plt.ylabel('Scale')
plt.colorbar(label='Magnitude')
plt.tight_layout()
plt.show()
```
